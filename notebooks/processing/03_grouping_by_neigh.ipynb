# 03 - Grouping by Neighborhood

Takes the `ALL_openings_closings.parquet` from notebook 01 and spatially aggregates business events to the neighborhood level. Uses functions.py file to sort by neighborhood and year, and separately by neighborhood, NAICS industry group, and year. Exports two parquets: `ALL_openings_closings_by_neighs_year.parquet` and `ALL_openings_closings_by_naics_neighs_year.parquet`.

```python
import sys
import geopandas as gpd

sys.path.insert(0, '../../src')
import functions as fx
```

```python
gdf_open_close = gpd.read_parquet('../../data/processed/ALL_openings_closings.parquet')
sf_neigh = gpd.read_file('../../data/processed/polygons/sf_neighborhoods.geojson')
```

```python
result = fx.group_points_by_poly_year(
    gdf_open_close,
    sf_neigh,
    id_col='neighborhood'
)
result.to_parquet('../../data/processed/ALL_openings_closings_by_neighs_year.parquet')
```

```python
result_naics = fx.group_points_by_poly_naics_year(
    gdf_open_close,
    sf_neigh,
    id_col='neighborhood'
)
result_naics.to_parquet('../../data/processed/ALL_openings_closings_by_naics_neighs_year.parquet')
```
